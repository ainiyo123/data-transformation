```python
import json
import re
import requests
import datetime
from bs4 import BeautifulSoup
import os
```

```python
htmlfile = open('covid19.html', 'r', encoding='utf-8')
htmlhandle = htmlfile.read()
soup = BeautifulSoup(htmlhandle, 'lxml')
tables = soup.find_all('table')
```

```python
import pandas as pd
import numpy as np
#Hubei
bs = BeautifulSoup(str(tables[1]),'lxml')
all_trs = bs.find_all('tr')

date = []
confirmed = []
death = []
cured = []

for tr in all_trs[1:]:
    all_tds = tr.find_all('td')
    if len(all_tds[0].text) > 5:
        t = time.strptime(all_tds[0].text,'%m月%d日%H时')
        date.append(str(t.tm_mon)+'/'+str(t.tm_mday))
    else:
        t = time.strptime(all_tds[0].text,'%m月%d日')
        date.append(str(t.tm_mon)+'/'+str(t.tm_mday))
    confirmed.append(all_tds[1].text)
    death.append(all_tds[2].text)
    cured.append(all_tds[3].text)

hb_df = pd.DataFrame(list(zip(date, confirmed, death, cured)), columns=['Date', 'Confirmed cases', 'Death', 'Cured'])
hb_df.to_csv('hb_df')
```
